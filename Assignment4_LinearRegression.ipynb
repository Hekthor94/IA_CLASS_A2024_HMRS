```python
#1. Load libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# 2. Load data.

california = fetch_california_housing()

df = pd.DataFrame(data=california.data, columns=california.feature_names)

df['Precio'] = california.target


# Select Indp. Variable

X = df[['MedInc']] # MedInc (Ingresos medios en la zona)

y = df['Precio']
```

```python
# 3. Data preprocessing

# Split the data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# 4. Model Selection and Training
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 5. Model Testing
# Predict the house prices on the test set
y_pred = model.predict(X_test)

# Calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
# 6. Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual Prices',alpha=0.6, edgecolor='k')
plt.plot(X_test, y_pred, color='red', label='Predicted Prices', linewidth=2)
plt.xlabel('Median Income (MedInc)')
plt.ylabel('House Price (Precio)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.title('Linear Regression: House Price Prediction based on Median Income')
plt.legend()
plt.show()
```

```python
# 6. Plot Training and Test Set Results
plt.figure(figsize=(10, 6))

# Plot training data (blue dots)
plt.scatter(X_train, y_train, color='blue', label='Training Data', alpha=0.6, edgecolor='k')

# Plot test data (green dots)
plt.scatter(X_test, y_test, color='green', label='Test Data', alpha=0.6, edgecolor='k')

# Plot predicted values for test data (red line)
plt.plot(X_test, y_pred, color='red', label='Predicted Prices (Test)', linewidth=2)

# Enhancing the plot aesthetics
plt.title('House Price Prediction: Training and Test Sets', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Median Income (MedInc)', fontsize=14, color='darkgreen')
plt.ylabel('House Price (Precio)', fontsize=14, color='darkgreen')

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper left', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 6. Plot Original Dataset with Test Predictions
plt.figure(figsize=(10, 6))

# Plot original dataset (all data points in blue)
plt.scatter(X, y, color='blue', label='Original Data', alpha=0.6, edgecolor='k')

# Plot predicted values for test data (red line)
plt.plot(X_test, y_pred, color='red', label='Predicted Prices (Test)', linewidth=2)

# Enhancing the plot aesthetics
plt.title('House Price Prediction: Original Data and Test Predictions', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Median Income (MedInc)', fontsize=14, color='darkgreen')
plt.ylabel('House Price (Precio)', fontsize=14, color='darkgreen')

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper left', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
