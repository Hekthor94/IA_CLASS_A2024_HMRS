```python
# Name: Hector Maximiliano Ramos Salazar
#Id: 2007690
# Day: Tuesday #Hour: N4-N6
#Text Analysis Tool
import string
from collections import Counter

def count_words(text):
    words = text.split()
    return len(words)

def count_sentences(text):
    sentences = text.split('.')
    return len([s for s in sentences if s.strip()])

def count_paragraphs(text):
    paragraphs = text.split('\n\n')
    return len([p for p in paragraphs if p.strip()])

def average_word_length(text):
    words = text.split()
    word_lengths = [len(word.strip(string.punctuation)) for word in words if word.strip(string.punctuation)]
    return sum(word_lengths) / len(word_lengths) if word_lengths else 0

def most_common_words(text, n=5):
    words = text.lower().split()
    stripped_words = [word.strip(string.punctuation) for word in words if word.strip(string.punctuation)]
    counter = Counter(stripped_words)
    return counter.most_common(n)

def main():
    # Read the file
    file_path = input("Enter the file path: ")
    try:
        with open(file_path, 'r') as file:
            text = file.read()
    except FileNotFoundError:
        print("File not found.")
        return

    # Perform analysis
    word_count = count_words(text)
    sentence_count = count_sentences(text)
    paragraph_count = count_paragraphs(text)
    avg_word_length = average_word_length(text)
    common_words = most_common_words(text)

    # Display results
    print(f"Total words: {word_count}")
    print(f"Total sentences: {sentence_count}")
    print(f"Total paragraphs: {paragraph_count}")
    print(f"Average word length: {avg_word_length:.2f}")
    print("Most common words:")
    for word, freq in common_words:
        print(f"{word}: {freq}")

#FILEPATH: /content/DONTREAME.txt
main()
```

```python
# Name: Hector Maximiliano Ramos Salazar
#Id: 2007690
# Day: Tuesday #Hour: N4-N6
#HANGMAN GAME
import random


def display_hangman(attempts):
    stages = [
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |     / \\
           -
        """,
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |     /
           -
        """,
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |
           -
        """,
        """
           --------
           |      |
           |      O
           |     \\|
           |      |
           |
           -
        """,
        """
           --------
           |      |
           |      O
           |      |
           |      |
           |
           -
        """,
        """
           --------
           |      |
           |      O
           |
           |
           |
           -
        """,
        """
           --------
           |
           |
           |
           |
           |
           -
        """
    ]
    return stages[attempts]

def hangmanGame():
    words = ['python', 'artificial', 'intelligence', 'computer', 'science', 'neuron']
    word = random.choice(words)
    guessed_word = ['_'] * len(word)
    attempts = 6
    guessed_letters = []

    print("HANGMAN: THE GAME")
    print(display_hangman(attempts))
    print(" ".join(guessed_word))

    while attempts > 0:
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.")
            continue

        guessed_letters.append(guess)

        if guess in word:
            for i, letter in enumerate(word):
                if letter == guess:
                    guessed_word[i] = letter
            print("Correct!")
        else:
            attempts -= 1
            print("Incorrect!")



        print("*************************")
        print(display_hangman(attempts))
        print(" ".join(guessed_word))

        if '_' not in guessed_word:
            print("Congratulations! You guessed the word!")
            break
    else:
        print(f"Game over! The word was '{word}'.")


hangmanGame()
```

```python
# Name: Hector Maximiliano Ramos Salazar
#Id: 2007690
# Day: Tuesday #Hour: N4-N6
#Number Guessing Game
import random

def numberGuessingGame():
    # Generate a random number between 1 and 100
    target_number = random.randint(1, 100)

    print("Number Guessing Game")
    print("Guess a number between 1 and 100.")

    while True:
        try:
            # Get the player's guess
            guess = int(input("Please enter your guess: "))

         #Check number
            if guess < target_number:
                print("Too low!")
            elif guess > target_number:
                print("Too high!")
            else:
                print(f"Congratulations! You've guessed the correct number: {target_number}")
                break
        except ValueError:
            print("Invalid input! Please enter an integer number.")


numberGuessingGame()
```
